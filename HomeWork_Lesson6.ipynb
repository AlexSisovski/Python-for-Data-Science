```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import load_boston # —Ç–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ
from sklearn.model_selection import train_test_split # —Ä–∞–∑–±–∏–≤–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä—ç–π–º—ã
from sklearn.linear_model import LinearRegression # —Å—Ç—Ä–æ–∏—Ç–º –º–æ–¥–µ–ª—å —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
from sklearn.metrics import mean_squared_error # —Å—Ä–µ–¥–Ω—è—è –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–∞—è –æ—à–∏–±–∫–∞
from sklearn.metrics import r2_score # –ø—Ä–∏–º–µ–Ω—è–µ–º –Ω–∞ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è—Ö
from sklearn.metrics import roc_auc_score # –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
```

## –ó–ê–î–ê–ù–ò–ï 1

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
data = boston['data']
target = boston['target']
feature_names = boston["feature_names"]
X = pd.DataFrame(data, columns=feature_names)
```

```python
X.info()
```

```python
X.shape
```

```python
X.head()
```

```python
y = pd.DataFrame(target, columns=["price"])
y.info()
```

```python
X.describe()
```

–†–∞–∑–±–µ–π—Ç–µ —ç—Ç–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–µ (X_train, y_train) –∏ —Ç–µ—Å—Ç–æ–≤—ã–µ (X_test, y_test) —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ train_test_split —Ç–∞–∫, —á—Ç–æ–±—ã —Ä–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ —Å–æ—Å—Ç–∞–≤–ª—è–ª 30% –æ—Ç –≤—Å–µ—Ö –¥–∞–Ω–Ω—ã—Ö, –ø—Ä–∏ —ç—Ç–æ–º –∞—Ä–≥—É–º–µ–Ω—Ç random_state –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Ä–∞–≤–µ–Ω 42. –°–æ–∑–¥–∞–π—Ç–µ –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –ø–æ–¥ –Ω–∞–∑–≤–∞–Ω–∏–µ–º lr —Å –ø–æ–º–æ—â—å—é –∫–ª–∞—Å—Å–∞ LinearRegression –∏–∑ –º–æ–¥—É–ª—è sklearn.linear_model. –û–±—É—á–∏—Ç–µ –º–æ–¥–µ–ª—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (–∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≤—Å–µ –ø—Ä–∏–∑–Ω–∞–∫–∏) –∏ —Å–¥–µ–ª–∞–π—Ç–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö. –í—ã—á–∏—Å–ª–∏—Ç–µ R2 –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ø—Ä–µ–¥–∫–∞–∑–∞–Ω–∏–π —Å –ø–æ–º–æ—â—å—é r2_score –∏–∑ –º–æ–¥—É–ª—è sklearn.metrics.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

–ú–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –ø–æ –æ–±—ä–µ–∫—Ç—É  ùë•=(ùë•1,‚Ä¶,ùë•ùëõ)  –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π, –∏—Å–ø–æ–ª—å–∑—É—è –ª–∏–Ω–µ–π–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é

ùëì(ùë•)=ùë§0+ùë§1‚ãÖùë•1+‚ãØ+ùë§ùëõ‚ãÖùë•ùëõ.
 
–ó–∞–¥–∞—á–∞ —Ç–∞–∫–æ–π –º–æ–¥–µ–ª–∏ - –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ –æ–±—É—á–µ–Ω–∏—è –ø–æ–¥–æ–±—Ä–∞—Ç—å —ç—Ç–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã  ùë§ùëñ  —Ç–∞–∫, —á—Ç–æ–±—ã –∑–Ω–∞—á–µ–Ω–∏–µ —ç—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –±—ã–ª–æ –∫–∞–∫ –º–æ–∂–Ω–æ –±–æ–ª–µ–µ –±–ª–∏–∑–∫–æ –∫ —Ä–µ–∞–ª—å–Ω–æ–º—É —Ü–µ–ª–µ–≤–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é  ùë¶  –æ–±—ä–µ–∫—Ç–∞  ùë• .

–ò—Ç–∞–∫, —á—Ç–æ–±—ã –æ–±—É—á–∏—Ç—å –º–æ–¥–µ–ª—å, –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ .fit, –≤ –∫–æ—Ç–æ—Ä—ã–π –ø–µ—Ä–µ–¥–∞—ë–º –Ω–∞—à—É —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –≤—ã–±–æ—Ä–∫—É:

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})
check_test.head()
```

```python
check_test["error"] = check_test["y_pred"] - check_test["y_test"]
check_test.head()
```

–ú–ï–¢–†–ò–ö–ò
–°—Ä–µ–¥–Ω—è—è –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–∞—è –æ—à–∏–±–∫–∞ –∏–ª–∏ mean squared error, —Å–æ–∫—Ä–∞—â—ë–Ω–Ω–æ mse. –û–Ω–∞ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –∫–∞–∫ —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—à–∏–±–æ–∫ –Ω–∞ –∫–∞–∂–¥–æ–º –∏–∑ –æ–±—ä–µ–∫—Ç–æ–≤

```python
mean_squared_error(check_test["y_pred"], check_test["y_test"])
```

```python
r2_score(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

pd.options.display.max_columns = 100
%matplotlib inline
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
df.shape
```

```python
df.info()
```

```python
 df.head(10)
```

```python
df.columns
```

```python
y = pd.Series(df['Class'])
X = df.drop(columns=['Class'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
parameters = [{'n_estimators': [10, 15], 
               'max_features': np.arange(3, 5),
               'max_depth': np.arange(4, 7)}]
model = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
model.fit(X_train,  y_train)
```

```python
model.best_params_
```

```python
result = model.predict_proba(X_test)
result.shape
```

```python
result
```

```python
print(result[0:,1])
```
